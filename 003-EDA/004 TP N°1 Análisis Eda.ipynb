```
plt.scatter(df2["production_cost_usd"], df2["revenue_usd"], alpha=0.3, s=10)

plt.title("Costo de Producción vs Ganancia en USD")
plt.xlabel("Costo de producción (USD)")
plt.ylabel("Ganancia (USD)")
plt.show()
```

Análisis del gráfico: Costo de Producción vs Ganancia

A simple vista se puede ver que la gran mayoría de los puntos que detallan las ganancias va de $0 a $200 USD, independientemente del costo de producción\. Esto indica que la mayor parte del contenido generado con IA no genera retorno económico acorde a la inversión para su producción\.
En cuanto a los Costos de producción, detallados en el eje X, se que confirma que las herramientas de IA permiten producir contenido a costos bajos donde la mayor concentración se da entre los $5 y $60 USD\.

Outliers
Hay una pequeña cantidad de puntos que se despegan hacia arriba, con ganancias que llegan hasta $2\.600 USD\. Lo llamativo es que estos outliers no se concentran en los costos más altos, sino que aparecen distribuidos a lo largo de todo el eje X, incluso en producciones de bajo costo \(menos de $20 USD\)\.

Conclusión

El costo de producción no es un predictor confiable de la ganancia\. El gráfico muestra que invertir más en producción no se traduce necesariamente en más retorno monetario\. Los contenidos más rentables no son los más costosos, sino que el rendimiento parece depender de otros factores como la plataforma, el tipo de contenido, el alcance orgánico o el engagement\. Esto es una señal positiva para creadores con presupuesto limitado: con herramientas de IA de bajo costo es perfectamente posible alcanzar ganancias altas, siempre que se optimicen las variables correctas\.

```

df3[['Financial Loss (in Million $)', 'Number of Affected Users', 'Incident Resolution Time (in Hours)']].hist(bins=25, figsize=(15, 4))
plt.suptitle('Distribucion de variables numericas')
plt.tight_layout()
plt.show()
 
```


Para esta sección del análisis, se eligió utilizar una serie de histogramas\. La razón principal para elegir este tipo de gráfico es que resulta ser la herramienta visual más adecuada para entender la distribución y el comportamiento general de variables numéricas continuas dentro de un gran volumen de datos \(en este caso, nuestro registro de amenazas cibernéticas\)\.

El gráfico visualiza tres variables clave de nuestro dataset:

Pérdida Financiera \(Financial Loss\): Medida en millones de dólares\.

Usuarios Afectados \(Number of Affected Users\): Cantidad de personas impactadas por incidente\.

Tiempo de Resolución \(Incident Resolution Time\): Medido en horas\.

¿Cómo se representan los datos en este gráfico?
El código empleado \(\.hist\(bins=25\)\) toma el rango total de valores de cada columna \(desde el mínimo hasta el máximo\) y lo divide en 25 intervalos o "bloques" iguales\.

En el eje horizontal \(X\), observamos estos rangos de valores \(por ejemplo, tramos de dinero, de usuarios o de horas\)\.

En el eje vertical \(Y\), observamos la frecuencia; es decir, la cantidad exacta de ataques cibernéticos que caen dentro de cada uno de esos intervalos\.

Conclusiones obtenidas de los datos:
Al observar la forma de los tres histogramas, se evidencia un patrón claro: los datos presentan una distribución uniforme\. Las barras mantienen alturas muy similares a lo largo de todos los rangos\. Esto nos indica que, dentro de nuestro registro, no existe un "ataque estándar" que predomine sobre el resto\. Las probabilidades de sufrir un ataque que genere bajas pérdidas o que se resuelva en pocas horas, son estadísticamente similares a las de enfrentar un ataque de altísimo costo financiero o que tome varios días en ser mitigado\.
